```python
#Count how many orders were shipped with Ship Mode = "First Class"
count=df[df['Ship Mode']=='First Class'].shape[0]
print("Number of orders with First Class shipping:", count)
```

```python
# Explanation:

# df['Ship Mode'] == 'First Class' â†’ creates a True/False mask for rows

# df[...] â†’ filters only those rows

# .shape[0] â†’ counts the number of rows (i.e., number of orders)
```

```python
# Count orders for all Ship Modes
ship_counts = df['Ship Mode'].value_counts()
print(ship_counts)
```
