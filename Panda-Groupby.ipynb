```python
#Pandas-Groupby practice
import pandas as pd
df=pd.read_csv(r"D:\Pandas-df\Global-Superstore.csv")
```

```python
#Q1. How many orders were shipped in each Ship Mode? or can you count rows per category?
df.groupby('Category').size()
```

```python
#Q2. How many unique customers are there in each Country?
df.groupby('Country').nunique()
```

```python
#Q3. What is the total Sales per Country?
df.groupby('Country')['Sales'].sum().reset_index(name='Total Sales')
```

```python
#Q4. What is the average Profit per Category?
df.groupby('Category')['Profit'].mean().reset_index(name="Avg Profit")
```

```python
#Q5. Find the maximum Discount given in each Region.
df.groupby('Region')['Discount'].max()
```

```python
#Q6. Get both the total and average Sales per Category. (Multiple aggregations are being tested here)
df.groupby('Category')['Sales'].agg(['sum','mean'])#here no paranthesis to sum n mean
```

```python
#6.1 #Q6. Get both the total and average Sales per Category. (Multiple aggregations are being tested here)
df.groupby('Category')['Sales'].agg(Total_Sales='sum', Average='mean')#here no paranthesis to sum n mean
#In Python, we use = instead of : when naming inside .agg().
```

```python
#Q7. Can you calculate total Sales, average Profit, and max Discount for each Segment?
df.groupby('Segment').agg({'Sales':'sum','Profit':'mean','Discount':'max'})
```

```python
#IN aggreagted style:
df.groupby('Segment').agg(
    Total_Sales=('Sales','sum'),
Total_Profit=('Profit','mean'),
Total_Discount=('Discount','max'))
```

```python
#Q9. Which Country has the highest total Sales?
df.groupby('Country')['Sales'].sum().sort_values(ascending=False).head(1)
#df.groupby('Country')['Sales'].sum().nlargest(1)
```

```python
df.groupby('Country')['Sales'].sum().nlargest(1)
```

```python
#Q10. Show only those Categories where the total Sales are greater than 500,000.
df.groupby('Category')['Sales'].sum().[df.groupby('Category')['Sales'].sum()>500,000]
```

```python
category_sales=df.groupby('Category')['Sales'].sum()
category_sales[category_sales>500000]
```

```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

# # Filter only categories with sales > 50,000
filtered_sales = category_sales[category_sales['Sales'] > 500000]

print(filtered_sales)
```
